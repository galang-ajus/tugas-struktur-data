```python
nilai_mahasiswa = []

for i in range(10):
    nilai = int(input(f"Input nilai ke-{i+1}: "))
    nilai_mahasiswa.append(nilai)

nilai_tertinggi = max(nilai_mahasiswa)
nilai_terendah = min(nilai_mahasiswa)

rata_rata = sum(nilai_mahasiswa) / len(nilai_mahasiswa)

jumlah_lulus = 0
for n in nilai_mahasiswa:
    if n >= 60:
        jumlah_lulus += 1

9print("\n===== HASIL PERHITUNGAN =====")
print("Data Nilai           :", nilai_mahasiswa)
print("Nilai Tertinggi      :", nilai_tertinggi)
print("Nilai Terendah       :", nilai_terendah)
print("Rata-rata Nilai      :", rata_rata)
print("Jumlah Mahasiswa Lulus :", jumlah_lulus)
```

```python
kategori_nilai = ["Nilai Tertinggi", "Nilai Terendah"]
data_nilai_grafik = [nilai_tertinggi, nilai_terendah]

plt.bar(kategori_nilai, data_nilai_grafik)

plt.title("Grafik Perbandingan Nilai Tertinggi dan Terendah")
plt.xlabel("Jenis Nilai")
plt.ylabel("Nilai")

plt.show()
```

```python


status = ["Lulus", "Tidak Lulus"]
jumlah = [jumlah_lulus, len(nilai_mahasiswa) - jumlah_lulus]

plt.bar(status, jumlah)

plt.title("Grafik Kelulusan Mahasiswa")
plt.xlabel("Status")
plt.ylabel("Jumlah Mahasiswa")

plt.show()
```
